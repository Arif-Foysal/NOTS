# NOTS-NIDS: Data Exploration
Explore CICIDS-2017 dataset: class distribution, feature statistics, and correlations.

```python
import sys, os
sys.path.insert(0, '..')
os.chdir('..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from config import Config

cfg = Config()
```

```python
# Load CICIDS-2017
from preprocessing.loader import load_cicids2017
df, label_col = load_cicids2017('data/cicids2017')
print(f'Shape: {df.shape}')
print(f'Label column: {label_col}')
df[label_col].value_counts()
```

```python
# Clean
from preprocessing.cleaner import clean_dataframe, encode_labels
df_clean, dropped = clean_dataframe(df, label_col)
print(f'Cleaned shape: {df_clean.shape}')
print(f'Dropped columns: {dropped}')

df_clean, label_map = encode_labels(df_clean, label_col)
print(f'Label map: {label_map}')
```

```python
# Class distribution plot
fig, ax = plt.subplots(figsize=(12, 5))
counts = df_clean[label_col].value_counts()
counts.plot(kind='bar', ax=ax, color=sns.color_palette('viridis', len(counts)))
ax.set_title('CICIDS-2017 Class Distribution')
ax.set_ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Feature statistics
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
feature_cols = [c for c in numeric_cols if c not in ['label_int', 'is_attack']]
print(f'Number of numeric features: {len(feature_cols)}')
df_clean[feature_cols].describe().T
```

```python
# Correlation heatmap (top 20 features by variance)
top_var = df_clean[feature_cols].var().nlargest(20).index.tolist()
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(df_clean[top_var].corr(), cmap='coolwarm', center=0, ax=ax)
ax.set_title('Feature Correlation (Top 20 by Variance)')
plt.tight_layout()
plt.show()
```
