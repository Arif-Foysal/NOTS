# NOTS-NIDS: Paper Figures
Reproduce all tables and figures for the IEEE T-IFS paper.

**Prerequisites:** Run `03_run_experiments.ipynb` first to generate `results/*.csv`.

```python
import sys, os
sys.path.insert(0, '..')
os.chdir('..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from config import Config
cfg = Config()

# Load saved results
exp1 = pd.read_csv(f'{cfg.RESULTS_DIR}/exp1_results.csv') if os.path.exists(f'{cfg.RESULTS_DIR}/exp1_results.csv') else None
exp2 = pd.read_csv(f'{cfg.RESULTS_DIR}/exp2_results.csv') if os.path.exists(f'{cfg.RESULTS_DIR}/exp2_results.csv') else None
exp3 = pd.read_csv(f'{cfg.RESULTS_DIR}/exp3_results.csv') if os.path.exists(f'{cfg.RESULTS_DIR}/exp3_results.csv') else None
exp4 = pd.read_csv(f'{cfg.RESULTS_DIR}/exp4_ablation.csv') if os.path.exists(f'{cfg.RESULTS_DIR}/exp4_ablation.csv') else None
exp5 = pd.read_csv(f'{cfg.RESULTS_DIR}/exp5_efficiency.csv') if os.path.exists(f'{cfg.RESULTS_DIR}/exp5_efficiency.csv') else None

print('Results loaded:')  
for name, df in [('Exp1', exp1), ('Exp2', exp2), ('Exp3', exp3), ('Exp4', exp4), ('Exp5', exp5)]:
    print(f'  {name}: {"OK" if df is not None else "MISSING"} {df.shape if df is not None else ""}')
```

```python
# Figure 1: Ablation study bar chart
from evaluation.plotter import plot_ablation_bars

if exp4 is not None:
    ablation_dict = {}
    for _, row in exp4.iterrows():
        ablation_dict[row['variant']] = {'DR': row['DR'], 'FPR': row['FPR']}
    plot_ablation_bars(ablation_dict, f'{cfg.FIGURES_DIR}/fig_ablation')
    print('Figure saved: fig_ablation.pdf/.png')
```

```python
# Figure 2: DR vs delta (Theorem validation)
from evaluation.plotter import plot_wasserstein_vs_delta

if exp2 is not None:
    epsilon_min = exp2['epsilon_min'].iloc[0]
    nots_results = {}
    for _, row in exp2.iterrows():
        nots_results[row['delta']] = {'DR': row['DR']}
    plot_wasserstein_vs_delta({'NOTS': nots_results}, epsilon_min, f'{cfg.FIGURES_DIR}/fig_dr_vs_delta')
    print('Figure saved: fig_dr_vs_delta.pdf/.png')
```

```python
# Figure 3: Efficiency table
from evaluation.plotter import plot_efficiency_table

if exp5 is not None:
    exp5_display = exp5.set_index('N')
    plot_efficiency_table(exp5_display, f'{cfg.FIGURES_DIR}/fig_efficiency')
    print('Figure saved: fig_efficiency.pdf/.png')
```

```python
# Summary Table: All experiments
print('\n' + '=' * 70)
print('RESULTS SUMMARY FOR IEEE T-IFS PAPER')
print('=' * 70)

if exp1 is not None:
    print('\nTable 1: Baseline Detection (Exp 1)')
    print(exp1.to_string(index=False))

if exp2 is not None:
    print('\nTable 2: White-box Attack (Exp 2)')
    print(exp2.to_string(index=False))

if exp3 is not None:
    print('\nTable 3: Black-box Attack (Exp 3)')
    print(exp3.to_string(index=False))

if exp4 is not None:
    print('\nTable 4: Ablation Study (Exp 4)')
    print(exp4.to_string(index=False))

if exp5 is not None:
    print('\nTable 5: Efficiency (Exp 5)')
    print(exp5.to_string(index=False))

print('\n' + '=' * 70)
print('All figures saved to:', cfg.FIGURES_DIR)
```
